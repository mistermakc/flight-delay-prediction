```python
# We create a column for each dataset to then concatenate both datasets.
df_19['YEAR'] = 2019
df_20['YEAR'] = 2020

# Let's see if the columns were created properly.
df_19.head()
```

```python
print('2019 dataset shape ' + str(df_19.shape))
print('2020 dataset shape ' + str(df_20.shape))

# We create a unique dataset
df = pd.concat([df_19,df_20])
print('Unique dataset shape ' + str(df.shape))
```

```python
# Getting rid of unnecessary columns
df = df.drop(['OP_CARRIER_AIRLINE_ID','OP_CARRIER',
                    'TAIL_NUM', 'ORIGIN_AIRPORT_ID','ORIGIN_AIRPORT_SEQ_ID',
                    'DEST_AIRPORT_ID','DEST_AIRPORT_SEQ_ID','Unnamed: 21'], axis=1)
df.head()
```

```python
cc = pd.read_csv('carrier_codes.csv', sep=";")

# Create a dictionary from the first and third columns
rename_dict = dict(zip(cc.iloc[:, 0], cc.iloc[:, 2]))

# Replace the values in the first column with the corresponding values from the dictionary
df["OP_UNIQUE_CARRIER"] = df["OP_UNIQUE_CARRIER"].replace(rename_dict)
df.rename(columns={"OP_UNIQUE_CARRIER": "AIRLINE"}, inplace=True)

df.head()
```

```python
# Let's change the index to be OP_CARRIER_FL_NUM (Flight number)
df.rename(columns={"OP_CARRIER_FL_NUM": "FLIGHT_NUMBER"}, inplace=True)
df = df.set_index('FLIGHT_NUMBER')
df.head()
```

```python
# Let's make a summary of the data
df_info = pd.DataFrame({'unique_values':df.nunique(),
              'missing_values': df.isna().sum(),
              'missing_values_perc': (df.isna().sum()/df.count()).round(2),
              'data_type':df.dtypes})

df_info.head(30)
```

```python
# We can see that there aren't that many missing values, so we can get rid of them.
# We remove rows with missing values.
df.dropna(inplace=True)
df.shape
```
