```python
# Checking if the data sets have the same columns.
print(list(data_2020.columns) == list(data_2019.columns))
```

```python
# We create a column for each dataset to then concatenate both datasets.
data_2019['year'] = 2019
data_2020['year'] = 2020

# Let's see if the columns were created properly.
data_2019.head()
```

```python
data_2020.head()
```

```python
print('2019 dataset shape ' + str(data_2019.shape))
print('2020 dataset shape ' + str(data_2020.shape))

# We create a unique dataset
data = pd.concat([data_2019,data_2020])
print('Unique dataset shape ' + str(data.shape))
```

```python
data.head()
```

```python
# Getting rid of unnecessary columns
data = data.drop(['OP_UNIQUE_CARRIER','OP_CARRIER_AIRLINE_ID','OP_CARRIER',
                    'TAIL_NUM', 'ORIGIN_AIRPORT_ID','ORIGIN_AIRPORT_SEQ_ID',
                    'DEST_AIRPORT_ID','DEST_AIRPORT_SEQ_ID','Unnamed: 21'], axis=1)
data
```

```python
# Let's change the index to be OP_CARRIER_FL_NUM (Flight number)
data = data.set_index('OP_CARRIER_FL_NUM')
data.head()
```

```python
data.info()
```

```python
# Let's make a summary of the data
data_info = pd.DataFrame({'unique_values':data.nunique(),
              'missing_values': data.isna().sum(),
              'missing_values_perc': data.isna().sum()/data.count(),
              'data_type':data.dtypes})

data_info
```

```python
# We can see that there aren't that many missing values, so we can get rid of them.
# We remove rows with missing values.
data.dropna(inplace=True)
data.shape
```

```python
data.head()
```

```python
# Now we want to transform the columns that are strings and make the categorical
# as well as variables with low unique values.
cols_to_cat = ['ORIGIN', 'DEST', 'DEP_DEL15', 'DEP_TIME_BLK', 'ARR_DEL15', 'CANCELLED', 'DIVERTED']

for col in cols_to_cat:
    data[col] = data[col].astype('category')
```
